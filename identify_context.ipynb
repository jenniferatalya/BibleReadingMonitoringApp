```python
import pandas as pd
from jaro import jaro_winkler_metric
```

```python
try:
    df = pd.read_csv("assets/classifiedchat_parse.csv")

    kitab = pd.read_csv("assets/biblebooknames.csv")
    kitab_list = kitab["Kitab"].tolist()

    def find_highest_score(book_names, input_text):
        max_score = 0
        current_kitab = ""

        for kata in book_names:
            similarity_score = jaro_winkler_metric(input_text, kata)
            if similarity_score > max_score:
                max_score = similarity_score
                current_kitab = kata

        return current_kitab

    context_list = []
    for index, row in df.iterrows():
        parse_text = row["Parse Text"]
        if pd.notnull(parse_text):
            print("Processing row", index, ": ", parse_text)
            current_kitab = find_highest_score(kitab_list, parse_text)
            context_list.append(current_kitab)
        else:
            context_list.append(None)


    df["context"] = context_list

    df.to_csv("assets/classifiedchat_context.csv", index=False)

    print("Data processing completed successfully.")

except Exception as e:
    print("An error occurred:", e)
```
