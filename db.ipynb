```python
import pandas as pd
```

```python
data = pd.read_csv('assets/biblechapters.csv')
data
```

```python
book = []
num = []
for line in data['Chapters']:
    tmp = line.split()
    if len(tmp) == 2:
        book.append(tmp[0])
        num.append(tmp[1])
    else:
        book.append(f'{tmp[0]} {tmp[1]}')
        num.append(tmp[-1])
```

```python
data['book'] = book
data['num'] = num
data
```

```python
len(data)
```

```python
sql_results = []
for i in range(len(data)):
    sql_results.append(f"""INSERT INTO chapter_table (chapter_name, book_name, number) VALUES ("{data['Chapters'][i]}", "{data['book'][i]}", {data['num'][i]});""")
```
