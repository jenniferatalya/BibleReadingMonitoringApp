```python
import pandas as pd
import re
import random
```

```python
def extract_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.readlines()

    result = []
    report_text = []
    for line in data:
        pattern = r'[\u200B-\u200F\u2028-\u202F\u205F-\u2064\uFEFF]'
        new_line = re.sub(pattern, ' ', line)
        new_line = new_line.replace('\n', '')
        if new_line.startswith('['):
            report_text.append(new_line)
        else:
            previous = report_text.pop(-1)
            report_text.append(previous + ' ' + new_line)

    pattern = r'\](.*?):'
    for line in report_text:
        temp = re.search(pattern, line)
        if temp:
            if temp not in result:
                result.append(temp.group(1))

    result_final = set(result)

    df = pd.DataFrame(result_final, columns=['Name'])
    
    return df
```

```python
df = extract_data('assets/chat.txt')
df
```
