```python
import pandas as pd

bible_chap = pd.read_csv('assets/biblechapters.csv')
bible_chap2 = pd.DataFrame(bible_chap)
print(bible_chap2)
```

```python
bible_dict = dict()

for i in range(len(bible_chap2)):
    bible_dict[bible_chap2['Chapters'][i]] = i

print(bible_dict)
```

```python
progress = [False]*len(bible_chap2)
```

```python
input_data = ["Kejadian 49", "Keluaran 3"]

def update_dictionary(input_data, dictionary, progress):
    start, end = input_data
    if start in dictionary and end in dictionary:
        start_chapter = dictionary[start]
        end_chapter = dictionary[end]
        for i in range(start_chapter, end_chapter+1):
            progress[i] = True
    else:
        print(f"Chapter does not exist in the dictionary.")
    return progress

print(update_dictionary(input_data, bible_dict, progress))
```
